```python
import pandas as pd
```

```python
tda = pd.read_csv('./2024C1_COMP230015_Técnicas_de_Diseño_de_Algoritmos.csv')
alg = pd.read_csv('./2024C1_COMP930005_Algoritmos_y_Estructuras_de_Datos_III.csv')
df = pd.concat([tda, alg], ignore_index=True)

# last word
df['Materia'] = df['data_turnos'].str.split().str[0]
df['Turno'] = df['data_turnos'].str.split().str[-1]
df['Categoria'] = ('20' +df['Lu'].str.split('/').str[-1]).astype(int)


# Definimos una función para extraer el año y la carrera
def extract_year_and_course(entry):
    # Dividimos la cadena por la coma para manejar entradas con múltiples carreras
    courses = entry.split(',')
    result = []
    for course in courses:
        # Encuentra la cadena dentro de los paréntesis finales
        year_course = course[course.rfind("(")+1:course.rfind(")")].split(' - ')
        # Separa el año y la cadena de la carrera
        # year = year_course[0].strip()
        career = year_course[1].strip()
        # result.append((year, career))
    return career

df['Plan'] = df['Carreras'].apply(extract_year_and_course)
```
